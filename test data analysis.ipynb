```python
def assign_letter(col = record["MTH 102"]):
    if col >= 70 and col < 100:
        result = 'A'
    elif col >= 60 and col < 70:
        result = 'B'
    elif col >= 50 and col < 60:
        result = 'C'
    elif col >= 40 and col < 50:
        result = 'D'
    else:
        result = 'F'
    return result

record['MTH 102'].apply(assign_letter)
```
